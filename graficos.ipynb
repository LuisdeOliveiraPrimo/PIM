# Graficos gerais

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# 1. Ler o arquivo JSON
with open('usuarios.json', 'r', encoding='utf-8') as f:
    dados = json.load(f)

# 2. Transformar os dados em DataFrame
df = pd.DataFrame(dados)

# 3. Exemplo 1 - Gráfico de barras de usuários por curso
curso_counts = df['curso'].value_counts()

plt.figure(figsize=(8, 5))
curso_counts.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Usuários por Curso')
plt.xlabel('Curso')
plt.ylabel('Quantidade de Usuários')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Exemplo 2 - Gráfico de barras de usuários por estado
estado_counts = df['estado'].value_counts()

plt.figure(figsize=(8, 5))
estado_counts.plot(kind='bar', color='green')
plt.title('Quantidade de Usuários por Estado')
plt.xlabel('Estado')
plt.ylabel('Quantidade de Usuários')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
